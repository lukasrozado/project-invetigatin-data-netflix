```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
print(netflix_df)
```

```python
netflix_subset = netflix_df[netflix_df['type'] == "Movie"]

print(netflix_subset)
```

```python
# Films from 90s

netflix_release = netflix_subset['release_year']
netflix_release1990 = netflix_subset[(netflix_release >= 1990) & (netflix_release < 2000)]

# Plot histogram for films released in the 90s
plt.hist(netflix_release1990['release_year'], bins=10)
plt.title('Films Released in the 90s')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
duration = netflix_release1990['duration'].mode()[0]
print(f"The most frequent movie duration is: {duration} minutes")
```

```python
# Find the most frequent movie duration
netflix_duration = netflix_release1990['duration']
netflix_duration_under_90 = netflix_release1990[netflix_release1990['duration'] < 90]
plt.hist(netflix_duration_under_90['duration'], bins=10)
plt.title('Movies with Duration Under 90 Minutes')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Count')
plt.show()
```

```python
# Find by genry

netflix_action = netflix_duration_under_90[(netflix_duration_under_90['genre'] == 'Action')]
short_movie_count = len(netflix_action)
print(netflix_action)
print(short_movie_count)
```
